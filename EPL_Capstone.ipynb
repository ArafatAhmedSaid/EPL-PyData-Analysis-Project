### 2. FEATURE ENGINEERING (Football Metrics)

#### a) Goal Features:

```python
df_match["GD"] = df_match["FTHG"] - df_match["FTAG"]
```

```python
df_match["Total Goals"] = df_match["FTHG"] + df_match["FTAG"]
```

```python
df_match[["HomeTeam", "AwayTeam", "FTHG", "FTAG", "GD","Total Goals"]].head()
```

#### b) Point From Results :

```python
df_match["Home_Points"] = np.where(df_match["FTR"] == "H", 3, np.where(df_match["FTR"] == "D", 1, 0))
```

```python
df_match["Away_Points"] = np.where(df_match["FTR"] == "A", 3, np.where(df_match["FTR"] == "D", 1, 0))
```

#### c) Team Points:

```python
home_points = df_match.groupby("HomeTeam")["Home_Points"].sum().sort_values(ascending=False)
```

```python
away_points = df_match.groupby("AwayTeam")["Away_Points"].sum().sort_values(ascending=False)
```

```python
team_points = home_points.add(away_points).sort_values(ascending=False)
team_points
```

```python
# Bar Plot:
plt.figure(figsize=(8, 5))

team_points.head(10).plot(kind="bar",color="black")

plt.title("Top 10 Teams by Total Points")
plt.xlabel("Team")
plt.ylabel("Points")

plt.show()
```

#### d) Total Goals Scored & Conceded:

```python
home_scored = df_match.groupby("HomeTeam")["FTHG"].sum()
away_scored = df_match.groupby("AwayTeam")["FTAG"].sum()
team_goals_scored = home_scored.add(away_scored, fill_value=0).sort_values(ascending=False)
team_goals_scored
```

```python
home_conceded = df_match.groupby("HomeTeam")["FTAG"].sum()
away_conceded = df_match.groupby("AwayTeam")["FTHG"].sum()
team_goals_conceded = home_conceded.add(away_conceded, fill_value=0).sort_values(ascending=False)
team_goals_conceded
```

#### e) Team Summary Table:

```python
team_table = pd.DataFrame({
    "Points": team_points,
    "Goals_Scored": team_goals_scored,
    "Goals_Conceded": team_goals_conceded
})
```

```python
team_table["Goal_Difference"] = (team_table["Goals_Scored"] - team_table["Goals_Conceded"])
```

```python
team_table.sort_values("Points", ascending=False)
```

#### f) Finishing & Contribution (Over/Under-Performance):

```python
# Finishing:

df_player["Finishing"] = df_player["Goals"] - df_player["xG"]
```

```python
df_player[["Player","Pos","Finishing"]].sort_values(by="Finishing",ascending=False).head(10)
```

```python
# Finishing Per 90:

df_player["Finishing"] = df_player["Goals per 90"] - df_player["xG per 90"]
```

```python
df_player[["Player","Pos","Finishing"]].sort_values(by="Finishing",ascending=False).head(10)
```

```python
# Contribution:

df_player["Contribution"] = df_player["Assists"] - df_player["xA"]
```

```python
df_player[["Player","Pos","Contribution"]].sort_values(by="Contribution",ascending=False).head(10)
```

```python
# Contribution Per 90:

df_player["Contribution"] = df_player["Assists per 90"] - df_player["xA per 90"]
```

```python
df_player[["Player","Pos","Contribution"]].sort_values(by="Contribution",ascending=False).head(10)
```

#### g) Team Quality:

```python
# Main Players:
df_player_main = df_player[df_player["Min Played"] >= 1000]
df_player_main.shape
```

```python
team_quality = df_player_main.groupby("Team").agg(
    Avg_GA_per90=("Goals + Assists per 90", "mean"),
    Avg_xGxA_per90=("xG+xA per 90", "mean"),
    Total_Finishing=("Finishing", "sum"))
```

### 3. EDA (Football Analysis Questions)

#### a) Chance Creation → Winning:

- Do teams with higher "Avg xG+xA per 90" earn more "Points"?

```python
corr_xGxA_pts = combined_team_profile["Avg_xGxA_per90"].corr(combined_team_profile["Points"])
corr_xGxA_pts
```

```python
x = combined_team_profile["Avg_xGxA_per90"].values
y = combined_team_profile["Points"].values

# Line of Best Fit:
m, c = np.polyfit(x, y, 1)
```

```python
# x,y
```

```python
# Scatter Plot:
plt.figure(figsize=(8, 6))
plt.scatter(combined_team_profile["Avg_xGxA_per90"], combined_team_profile["Points"])
plt.title("Team Points vs Avg xG+xA per 90")
plt.xlabel("Avg xG+xA per 90")
plt.ylabel("Points")

# Linear Regression Line:
plt.plot(x, m * x + c)

plt.show()
```

#### b) Over/Under-Achievers :

- Which teams “beat” their Squad Quality? ("Lucky Teams" vs "Wasteful Teams")

```python
combined_team_profile["Expected_Points_from_xGxA"] = m * combined_team_profile["Avg_xGxA_per90"] + c
```

```python
combined_team_profile["Points_OverUnder"] = combined_team_profile["Points"] - combined_team_profile["Expected_Points_from_xGxA"]
```

```python
# combined_team_profile
```

```python
team_overunder = combined_team_profile.sort_values("Points_OverUnder", ascending=False)[["Points","Expected_Points_from_xGxA","Points_OverUnder"]]
```

```python
colors = ['green' if j > 0 else 'red' for j in team_overunder["Points_OverUnder"]]

# Bar Plot:
plt.figure(figsize=(10, 5))
plt.bar(team_overunder.index, team_overunder["Points_OverUnder"], color=colors)
plt.title("Team Over/Under Performance (Actual Points - Expected Points)")
plt.xlabel("Team")
plt.ylabel("Points Over/Under")
plt.xticks(rotation=90)

plt.show()
```

### 5. INSIGHTS

KEY POINTS TO NOTE:

- Chance Quality is the best predictor of league success:- Teams that consistently generate and take chances per 90 win more points.

- Over-performers:- Some teams gained an edge through structure (Non-chaotic matches), game management, or performance in tight margins ~ "Luck".

- Defense is the next best predictor of league success.

- Finishing helps teams get points overall ~ 'Average' predictor of success.

- SOT dominance is also a short-term predictor of match success.

CONCLUSION:

SUCCESS in EPL is mainly based on CHANCE QUALITY and DEFENSIVE STRENGTH. Overachievement depends on how a team is STRUCTURED and how it controls games. Teams should also focus on SHOT QUALITY and FINAL 3RD DECISION-MAKING  in training.

RECOMMENDATIONS:

- Prioritize defensive structure as foundation of success.

- Improve shot quality rather than shot volume.

- Focus on game management in tight matches.

- Target players with high xG+xA per 90 rather than total goals during recruitment.
